# **Split de PoKi Arabe**

```python
import pandas as pd
import numpy as np
import os
from zipfile import ZipFile

df = pd.read_csv("/kaggle/input/poki-arabic/pokir_dataset_arabe.csv") 

n_parts = 61 # So that each one contains only 1000


chunks = np.array_split(df, n_parts)
output_folder = "/kaggle/working/poki-arabe"
os.makedirs(output_folder, exist_ok=True)


for i, chunk in enumerate(chunks, start=1):
    chunk.to_csv(f"{output_folder}/poki-arabe-{i}.csv", index=False)

# Zip all CSVs into one file
zip_filename = "/kaggle/working/poki-arabe.zip"
with ZipFile(zip_filename, 'w') as zipf:
    for csv_file in os.listdir(output_folder):
        csv_path = os.path.join(output_folder, csv_file)
        zipf.write(csv_path, arcname=csv_file)

print(f"Done! All files are zipped here: {zip_filename}")
```
